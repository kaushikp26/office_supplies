```python
import pandas as pd
df=pd.read_csv('office_supplies.csv',index_col='Order Date')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isna().sum()
df.dropna(inplace=True)
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
plt.figure(figsize=(9.32,4.32))
sns.set_theme(style='white')
sns.boxplot(data=df,color='b',whis=1.5)
```

```python
plt.figure(figsize=(10,8.4))
sns.heatmap(df.corr(),vmin=-1,vmax=1,center=0,cmap='crest',annot=True)
```

```python
products=df['Product Name'].value_counts(normalize=False)
products
```

```python
sns.countplot('Region',hue='Category',data=df)
plt.show()
sns.countplot('Region',hue='Sub-Category',data=df)
plt.legend(bbox_to_anchor=(1.02, 1),loc='best')
plt.show()
```

```python
sns.jointplot('Region','Sales',data=df)
```

```python
sns.pairplot(df,hue='Region')
```

```python
df['Year'] = pd.DatetimeIndex(df.index).year.astype(int)
df['Month'] = pd.DatetimeIndex(df.index).month
df.head()
```

```python
sns.barplot('Region','Sales',hue='Category',data=df)
plt.title('Sales: Region Level')
plt.show()
sns.barplot('Region','Profit',hue='Category',data=df)
plt.title('Profit: Region Level')
plt.show()
```

```python
south=df[df['Region']=='South']
central=df[df['Region']=='Central']
west=df[df['Region']=='West']
east=df[df['Region']=='East']
```

```python
def charts(d):
    sns.countplot('Category',data=d)
    plt.title('Amount of Supplies in Region: Category level')
    plt.show()
    sns.countplot('Sub-Category',data=d)
    plt.xticks(rotation=90)
    plt.title('Amount of Supplies in Region: Sub-Category level')
    plt.show()
    sns.lineplot('Month','Sales',data=d,hue='Category',ci=None)
    plt.title('Trend of Sales each Month:Category level') 
    plt.figure(figsize=[20,20])
    plt.show()
    sns.lineplot(data=d,x='Year',y='Sales',hue='Category',ci=None)
    plt.title('Trend of Sales each Year:Category level')
    plt.figure(figsize=[20,20])
    plt.show()
    sns.lineplot(data=d,x='Year',y='Profit',hue='Category',ci=None)
    plt.title('Trend of Profit each Year:Category level')
```

```python
charts(south)
```

```python
charts(central)
```

```python
charts(west)
```

```python
charts(east)
```
